We can replace the missing age values with the median.

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_median'] = titanic['Age'].fillna(titanic['Age'].median())
titanic.head(7)
```

We can replace the missing age values with the median separately for males and females.

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_median'] = titanic['Age'].fillna(titanic.groupby('Sex')['Age'].transform("median"))
titanic.head(20)
```

For categorical variables the missing values could be replaced with the mode.

```python
titanic['Embarked'].mode()
```

```python
titanic['Embarked'].mode()[0]
```

```python
titanic["Embarked_Mode"] = titanic["Embarked"].fillna(titanic['Embarked'].mode()[0])
titanic[titanic['Embarked'].isna()]
```

# Lambda

In Python, an anonymous function is a function that is defined without a name. While normal functions are defined using the **def** keyword in Python, anonymous functions are defined using the **lambda** keyword. Hence, anonymous functions are also called lambda functions.

    Syntax of Lambda Function in python
    lambda arguments: expression
    
Lambda functions can have any number of arguments but only one expression. The expression is evaluated and returned. 

We use lambda functions when we require a nameless function for a short period of time.

In Python, we generally use it as an argument to a higher-order function (a function that takes in other functions as arguments). 

```python
add_qm = lambda a: a + '???'
```

```python
add_qm('what')
```

```python
raise_to_power = lambda x, y: x ** y
```

```python
raise_to_power (2, 3)
```

---------------------------------

```python
titanic['Alone2'] = titanic['FamilySize'].apply(lambda x: 1 if x==0 else 0)
#a new binary variable that takes the value 1 if the passenger was travelling alone, 0 if not
titanic.head()
```

```python
titanic['Adult'] = titanic['Age'].apply(lambda x: 1 if x>=18 else 0)
#binary variable that takes the value 1 if the passenger was an adult, 0 if not
titanic.head(20)
```

# Function

In Python, a function is a group of related statements that performs a specific task.

A function has a name to uniquely identify it and parameters (arguments) through which we pass values to a function. 

```python
def get_age_gr(x):
    if x < 21:
        return 'young'
    elif x >=21 and x<60:
        return 'adult'
    elif x>=60:
        return 'old'
#we do not use else to prevent Nan values from entering group 'old'
```

-----------------------------------

```python
titanic['Age_groups'] = titanic['Age'].apply(get_age_gr)
#create a categorical variable based on a quantitative
titanic.head(20)
```

## 6. Ranging

```python
titanic = pd.read_csv('titanic.csv', index_col=0)
titanic['Age_rank_1'] = titanic['Age'].rank()
titanic.sort_values(by='Age').head(15) 
#the youngest passenger gets rank 1
#in the case of tied ranks (two or more observations have the same value) the average rank is calculated
```

```python
titanic['Age_rank_2'] = titanic['Age'].rank(method='max') 
titanic.sort_values(by='Age').head(15)
#the youngest passenger gets rank 1
#in the case of tied ranks the maximum rank is calculated
```

```python
titanic['Age_rank_3'] = titanic['Age'].rank(method='min') 
titanic.sort_values(by='Age').head(15)
#the youngest passenger gets rank 1
#in the case of tied ranks the maximum rank is calculated
```

```python
titanic['Age_rank_4'] = titanic['Age'].rank(method='min', na_option='bottom') 
titanic.head(19)
#the youngest passenger gets rank 1
#in the case of tied ranks the maximum rank is calculated
#if the age is not known, the observation is at the end of the rating
```

```python
titanic['Age_rank_5'] = titanic['Age'].rank(ascending=False) 
titanic.sort_values(by='Age', ascending=False).head(15)
#the oldest passenger gets rank 1
```
